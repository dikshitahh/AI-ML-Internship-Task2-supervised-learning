#Linear Regression
The main objective of this experiment to implement Linear Regression using Scikit-learn and understands how does it predicts continuous numeric values.

Linear regression is a supervised machine learning algorithm used for predicting continuous values. It establishes alinear relationship between input features and the target variable.

#Applications
- House Price Prediction
- Salary Prediction
- Sales Forecasting
- Temperature Prediction

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns = housing.feature_names)
df['Price'] = housing.target
df.head()
```

#Dataset Description
The California Housing dataset contains information about houses in California.

#Features
- MedInc: Median Income
- HouseAge: House Age
- AveRooms: Average Rooms
- AveBedrms: Average Bedrooms
- Population: Population
- AveOccup: Average Occupancy
- Latitude
- Longitude

#Target Variable
Price of the house

```python
from sklearn.model_selection import train_test_split
x = df.drop('Price' , axis = 1)
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2,random_state= 42)
```

#Data Splitting
The Dataset is divided into two parts:
- Training Data (80%)
- Testing Data (20%)

Training Data is used to train the model while Testing Data evaluates its performance.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

#Model Training
The Linear Regression model is trained using the training dataset.

The model learns the relationship between input features and house prices.

```python
predictions = model.predict(X_test)
predictions[:5]
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print("Mean Absolute Error:", mae)
print("R2 Score", r2)
```

# Performance Evaluation

##Mean Absolute Error(MAE):
It measures the average prediction error.

##R2 score:
It measures how well the model explains the variation in data.

Higher R2 values indicate better model performance.

#Conclusion
In this experiment, Linear Regression was implemented using the california housing datasets. The model was trained and evaluated using Mean Absolute Error(MAE) and R2 Score.
The result demonstrate that Linear Regression can effectively predict continuous numerical values and is suitable for regression problems.
